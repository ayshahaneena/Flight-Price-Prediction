### Feature Engineering

```python
# converting to datetime 
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
```

```python
df.info()
```

```python
# Creating new features of date , month and year
df['date'] = df['Date_of_Journey'].dt.day
df['month'] = df['Date_of_Journey'].dt.month
df['year'] = df['Date_of_Journey'].dt.year
```

```python
df.head(2)
```

```python
# checking the new features date , month , year in int format
df.info()
```

 * we no longer need this date_of_journey , so lets drop it

```python
# deleting the date_of_journey column
df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
df.head(2)
```

 * In the Arrival_Time column , we only require the time . Date should be removed

```python
df.info()
```

```python
# extracting only time
df['Arrival_Time']= df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
df.head()
```

```python
# spliting time into hour and minutes
df['Arrival_hour']= df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']= df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df.info()
```

```python
# changing hour and minutes format from object to int
df['Arrival_hour']= df['Arrival_hour'].astype(int)
df['Arrival_min']= df['Arrival_min'].astype(int)
```

```python
df.info()
```

```python
# we no longer need the Arrival_Time column , drop it 
df.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
df.head()
```

 * make it int format
